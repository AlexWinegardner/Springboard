# Machine Learning Modeling for Astrophysical Origin Classification

The purpose of this notebook is to build a machine learning model that is able to predict whether a detected signal will have a very high probability of being a real astrophysical event (>= 99%) based just on the value of its matched filter singal-to-noise ratio (SNR).

Earlier in our EDA notebook we discovered a very steep drop-off in the astrophysical origin probability for events with an SNR less than ~10 (image included below). Instead of just manually picking out any particular SNR value as a decision boundary, we will build three distinct classification models and compare how well they would be expected to perform when classifying new data. 

The reason for only using SNR as a single feature is to test how much predictive power it alone has over the astrophysical origin probability of an event. Of course, consideringing other features from our dataset might also help improve the accuracy of the models, but that might reduce the utility that a simpler model with just SNR might provide. 

Including other features would require that the full parameter estimations for each event to have already been done, but the utility of the model would be to serve as an initial screening to determine if it is even worth it to spend the time and computational resources in calculating the rest of the parameters. As gravitational wave detectors become more sensitive and are able to pick up signals at a much higher rate, future observation runs might need to prioritize which signals to estimate parameters for first.

## Loading modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import time
```

## Data Pre-processing recap

```python
df = pd.read_csv('GWTC_preprocessed.csv')
```

```python
X = df[['network_matched_filter_snr']] 
y = df['p_astro_binary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Modeling

We will create three models: logistic regression, decision tree, and support vector machines (SVM). Each one will then be compared through an analysis of their respective accuracy, precision, and recall scores, as well as cross-validation. Once a model has been chosen as the best one, we will then use hyperparameter tuning to see if we can make it even better.

### Logistic Regression

```python
# Training the logistic regression model with the binarized target
logreg_model = LogisticRegression()
logreg_model.fit(X_train, y_train)

# Predicting on the test set
y_pred = logreg_model.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_result = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print('Accuracy:', round(accuracy, 2) * 100, '%', '\n')
print('Classification Report:', '\n\n', classification_report_result)
```

We can see right away that a simple model such as logistic regression is able to get very high scores across the board. This is a promising start and corroborates our intuition that SNR alone would be good predictor. 

```python
# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay(conf_matrix, display_labels=['Class 0', 'Class 1']).plot(values_format='d', cmap='Blues', ax=plt.gca())
plt.title('Confusion Matrix')
plt.show()
```

Above we can visualize the confusion matrix and see that only one of the labels were predicted incorrectly. This was for a class 1 label (p_astro >= 0.99) being mistaken as a class 0 label (p_astro <= 0.99). This probably happened to an event that had its SNR value very close to the decision boundary.

For class 0, a precision of 0.92 means that 92% of the predicted labels were classified correctly (12 out of a total of 13 predictions), while a recall score of 1.00 means that 100% of the true labels were classified correctly (12 out of 12). 

For class 1, a precision of 1.00 means that 100% of the predicted labels were classified correctly (14 out of 14 predictions), while a recall score of 0.93 means that 93% of the true labels were classified correctly (14 out of a total of 15).

### Decision Tree

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
classification_report_result = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print('Accuracy:', round(accuracy, 2) * 100, '%', '\n')
print('Classification Report:', '\n\n', classification_report_result)

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay(conf_matrix, display_labels=['Class 0', 'Class 1']).plot(values_format='d', cmap='Blues', ax=plt.gca())
plt.title('Confusion Matrix')
plt.show()
```

Our decision tree model performed slightly worse, this time incorrectly predicting two labels. While decision trees are prone to overfit (and this might be the reason the performance got worse), they are at least more interpretable. Below I include a graph showing how the decision tree model reached its results:

```python
# Plotting decision tree graph
plt.figure(figsize=(12, 14))
plot_tree(dt_model, filled=True, feature_names=['snr'], class_names=['0', '1'])
plt.title('Decision Tree for Astrophysical Origin Prediction')
plt.show()
```

It is interesting to see that it picked out an SNR of 9.85 as its first split, which is close to the value of ~10 we could see from the plot at the beginning of the notebook.

### SVM

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
classification_report_result = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print('Accuracy:', round(accuracy, 2) * 100, '%', '\n')
print('Classification Report:', '\n\n', classification_report_result)

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay(conf_matrix, display_labels=['Class 0', 'Class 1']).plot(values_format='d', cmap='Blues', ax=plt.gca())
plt.title('Confusion Matrix')
plt.show()
```

This time we got a perfect score, but sometimes that may be cause for concern. In order to account for this, we will perform cross-validation on all our models, which will hopefully give a more representative expectation of how they would perform on new, unseen data:

## Cross-Validation

```python
models = [logreg_model, dt_model, svm_model]
model_names = ['Logistic Regression', 'Decision Tree', 'SVM']

for model, model_name in zip(models, model_names):
    cv_results = cross_validate(model, X, y, cv=5, scoring=['accuracy', 'precision', 'recall'])
    
    accuracy_mean = cv_results['test_accuracy'].mean()
    accuracy_std = cv_results['test_accuracy'].std()
    
    precision_mean = cv_results['test_precision'].mean()
    precision_std = cv_results['test_precision'].std()
    
    recall_mean = cv_results['test_recall'].mean()
    recall_std = cv_results['test_recall'].std()
    
    print(f"{model_name}\n"
          f"Accuracy: {accuracy_mean:.1%} (±{accuracy_std:.1%})\n"
          f"Precision: {precision_mean:.1%} (±{precision_std:.1%})\n"
          f"Recall: {recall_mean:.1%} (±{recall_std:.1%})\n")
```

This seems more realistic, before we might have just happened to have been making predictions on a "lucky" test set. However, it is still apparent that the SVM model performs better across all key metrics. We will select it as the best model and continue improving it through hyperparameter tuning:

## Hyperparameter Tuning

```python
param_grid_svm = {
    'C': [0.1, 1, 10],  # Regularization parameter
    'gamma': ['scale', 0.1, 1, 10],  # Kernel coefficient
    'kernel': ['rbf', 'linear', 'poly']  # Specifies the kernel type to be used in the algorithm
}

# Creating a GridSearchCV object for SVM
grid_search_svm = GridSearchCV(SVC(random_state=42), param_grid_svm, cv=5, scoring='accuracy')

# Starting a timer to see how long it takes to fit the model
start_time = time.time()

# Fitting the model
grid_search_svm.fit(X, y)

# End of timer
end_time = time.time()

# Calculating the duration
duration = end_time - start_time

# Best parameters
best_params_svm = grid_search_svm.best_params_

print("Best Parameters:", best_params_svm)
print("Duration:", duration, "seconds")
```

That didn't take very long, given that our dataset is small.

Now that we have the best parameters for our SVM model, we will build a new version and compare it to the original:

```python
svm_model_tuned = SVC(C = 1, gamma = 'scale', kernel = 'linear') 

models = [svm_model, svm_model_tuned]
model_names = ['SVM', 'SVM_Tuned']

for model, model_name in zip(models, model_names):
    cv_results = cross_validate(model, X, y, cv=5, scoring=['accuracy', 'precision', 'recall'])
    
    accuracy_mean = cv_results['test_accuracy'].mean()
    accuracy_std = cv_results['test_accuracy'].std()
    
    precision_mean = cv_results['test_precision'].mean()
    precision_std = cv_results['test_precision'].std()
    
    recall_mean = cv_results['test_recall'].mean()
    recall_std = cv_results['test_recall'].std()
    
    print(f"{model_name}\n"
          f"Accuracy: {accuracy_mean:.1%} (±{accuracy_std:.1%})\n"
          f"Precision: {precision_mean:.1%} (±{precision_std:.1%})\n"
          f"Recall: {recall_mean:.1%} (±{recall_std:.1%})\n")
```

We can see a clear improvement across all key metrics. Particularly, we notice the tuned model has gained an extra percent for accuracy and almost two percent for recall (while very slightly improving precision).

## Conclusion

We have successfully built a machine learning model that helps us accomplish our goal of predicting whether a signal will have a very high astrophysical origin probability (>= 99%) based just on its SNR. 

Our highest score was for recall which in this case is what we want – it means that the vast majority (96.4% on average) of labels with p_astro >= 99% have been correctly classified. The lower precision (91.6% on average) means that some labels with p_astro <= 99% were predicted to have p_astro >= 99%. However, even for these cases, it is probable they still had pretty high p_astro values nonetheless (it is much more likely for events with p_astro between 95-98% to be misclassified than for events with p_astro between 60-80%, for example). 

It is still to be seen if implementing a model like this before proceeding with the full parameter estimations is actually useful or practical, especially now while gravitational wave detections are relatively few. But with future observation runs and upgrades to detector sensitivities, if there are more potential signals coming in than what is computationally feasible to keep up with, then having a method similar to what we have explored here as an early filter might prove valuable. 

## Future Improvements

While this notebook has served as a demonstration of what could be possible by making predictions based just on SNR, if there is ever serious interest in pursuing it further there are several things which can be done to improve. 

- The first is to simply use more data once it's available. The more data the better!
- The second is to consider even more models, such as gradient boosting machines (GBM) / XGBoost, and to perform an even more exhaustive set of hyperparameter tuning for all models.
- The third is to determine if there is another feature that is not depedent on the results of full parameter estimations which could be used in conjuction with SNR for early screening.

Implementing the steps above could help build an even better model for handling new, unseen data. However, it is important to always remember the limitations of the model. When calculating the astrophysical probability, several other factors are taken into account, and a machine learning model alone should not be expected to authoritatively classify everything correctly. To be useful, having a human expert in the loop for validation may be necessary as well as including other analysis methods in conjuction. If successful, the model could potentially help create a catalog of very high confidence events (astrophysical origin probability >= 99%) in a more efficient manner.
